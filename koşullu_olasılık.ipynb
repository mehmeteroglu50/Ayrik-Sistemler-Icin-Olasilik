```python
# Bayes Teoremi'ni kullanarak çözüm
p_B_given_A = 0.80  # Hasta olan bir çocuğun test sonucunun pozitif çıkma olasılığı
p_A = 0.01  # Çocuğun hastalığa yakalanma olasılığı
p_not_A = 1 - p_A  # Çocuğun hastalığa yakalanmama olasılığı
p_B_given_not_A = 0.10  # Hasta olmayan bir çocuğun test sonucunun pozitif çıkma olasılığı

# Bayes Teoremi'ni kullanarak çocuğun gerçekten hasta olma olasılığını hesaplama
p_A_given_B = (p_B_given_A * p_A) / ((p_B_given_A * p_A) + (p_B_given_not_A * p_not_A))

print("Çocuğun gerçekten hasta olma olasılığı:", p_A_given_B)
```

```python
# Bayes Teoremi'ni kullanarak çözüm
p_B_given_A = 0.5  # Kız çocuğun saklambaç oynamak istememe olasılığı
p_A = 7 / 12  # Çocuğun kız olma olasılığı
p_B = 8 / 12  # Rastgele seçilen çocuğun saklambaç oynamak istememe olasılığı

# Bayes Teoremi'ni kullanarak çocuğun kız olma olasılığını hesaplama
p_A_given_B = (p_B_given_A * p_A) / p_B

print("Çocuğun kız olma olasılığı:", p_A_given_B)

```
